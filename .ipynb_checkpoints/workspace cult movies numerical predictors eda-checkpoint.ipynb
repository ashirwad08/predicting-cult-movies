```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pprint import pprint
import pickle
import re
%matplotlib inline
```

```python
with open(r"movie_DF_0428.p", "rb") as input_file:
    mov = pickle.load(input_file)
```

```python
mov.shape

#remove runtime_mins Nans
mov=mov[mov.runtime_mins.notnull()]
```

```python
#get representative sample from noncult labels to compare
cult = mov[mov.isCult==1]
nocultSample = mov[mov.isCult==0].sample(432)

movsample = pd.concat([cult, nocultSample])
#cult.isCult.sum()
movsample.shape
```

```python
#numeric predictors only
num_movsample = movsample.loc[:,['prod_budget_ADJ','CULT_INDEX','isCult']]
```

```python
print(mov.sort_values(by='CULT_INDEX', ascending=False).head(50))
```

```python
#sum(num_movsample.num_theaters.isnull())
cult.sample(5)
```

```python
#scale! too lazy to type, using sci-kitlearn
from sklearn.preprocessing import MinMaxScaler
num_movsample.loc[:,['prod_budget_ADJ']]=num_movsample.loc[:,['runtime_mins','prod_budget_ADJ']].apply(lambda row:MinMaxScaler().fit_transform(row))
```

```python
#sns.pairplot(movsample,dropna=True,hue='isCult', size=5)
sns.set_style('whitegrid')
grid = sns.FacetGrid(movsample, row='genre', col='isCult', size=5, ylim=(0,0.02),sharey=True)
grid.map(sns.regplot,'prod_budget_ADJ','CULT_INDEX',fit_reg=False)
```
